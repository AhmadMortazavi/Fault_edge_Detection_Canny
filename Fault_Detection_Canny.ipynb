# Read siesmic and variance data

```python
import pandas as pd
variance=pd.read_csv(r'D:\ML\Ahmad codes\IL_314_Fault_Var.csv')
seismic=pd.read_csv(r'D:\ML\Ahmad codes\IL_314_Fault.csv')
```

```python
variance=variance.values
```

```python
plt.imshow(seismic, cmap="gray")
plt.title('Seismic')
plt.xlabel('CRP Number')
plt.ylabel('Time sample Number')
plt.show()
```

```python
plt.imshow(variance, cmap="gray")
plt.title('variance')
plt.xlabel('CRP Number')
plt.ylabel('Time sample Number')
plt.show()
```

```python
variance.dtype
```

```python
# convert float 64 to int8, canny only works with int 8
variance = variance / variance.max() #normalizes data in range 0 - 255
variance = 255 * variance
variance8 = variance.astype(np.uint8)
plt.imshow(variance8,cmap="gray")
plt.title('variance 8 bit')
plt.xlabel('CRP Number')
plt.ylabel('Time sample Number')
plt.show()
```

```python
variance8.dtype
```

```python
edges_canny = cv2.Canny(variance8, threshold1=40, threshold2=100)
plt.imshow(edges_canny,cmap="gray")
plt.title('Fault detection using Canny')
plt.xlabel('CRP Number')
plt.ylabel('Time sample Number')
plt.show()
```
